```python
def calculate_zscore(price_data, market):
            spread_series = price_data[market]
            mean = spread_series.rolling(window=200).mean()
            std = spread_series.rolling(window=200).std()
            price_data[f'z_score_{market}'] = (spread_series - mean) / std

price_data = pd.read_csv('./data_15m.csv')

for market in price_data:
    if market == 'time':
        continue
    calculate_zscore(price_data, market)
price_data.to_csv('zscore_15m.csv')
```

```python
price_data = pd.read_csv('./data_1H.csv')

for market in price_data:
    if market == 'time':
        continue
    calculate_zscore(price_data, market)
price_data.to_csv('zscore_1H.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=price_data['time'], y=price_data['z_score_ETHUSDT'], data=price_data, marker='o')

# Customize the plot
plt.title('Time Series Plot of Value')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create the KDE plot
plt.figure(figsize=(10, 6))
sns.kdeplot(price_data['z_score_ETHUSDT'], fill=True)

# Customize the plot
plt.title('KDE Plot of Value')
plt.xlabel('Value')
plt.ylabel('Density')
plt.tight_layout()

# Show the plot
plt.show()
```
